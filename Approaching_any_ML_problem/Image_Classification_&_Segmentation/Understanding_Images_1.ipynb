<i>Image is nothing but a matrix of numbers.</i><br>
A <b>grayscale image</b> is a 2-D matrix with values ranging from 0 to 255.<br>
0 in black and 255 is white and in between you have all shades of grey.

```python
import numpy as np
import matplotlib.pyplot as plt
random_image = np.random.randint(0,256,(256,256))
plt.figure(figsize=(5,5))
plt.imshow(random_image,cmap='gray',vmin=0,vmax=255)
plt.show()
```
